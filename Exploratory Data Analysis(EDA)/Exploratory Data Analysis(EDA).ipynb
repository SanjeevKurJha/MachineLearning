## Correlation analysis and feature selection of data 

####  in case if within independent variable,This will explain you how the under line data and feature intrect with each other like more posative value between the feature will be highly correleted(we may leave it) and more neagative value between feature will be less correleted (we need to consider it)

#### in case with independent variable value with  dependent varibale then we need to consider highly correlated(either negative or posative) value to check the impect

#### Allthough this is not usefull for our current dataset

```python
data_Frame.corr()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data_Frame.corr(),annot=True)
plt.show()
```

#### Stronger relationships in are represented by values in the heat map that are closer to negative or positive one
#### if values are closer to zeor then its Weaker relationships
#### Positively related variables, values with correlations between zero and positive one, indicate that one variable increases as the other variable increases
#### Negatively related variables, values with correlations between negative one and zero, indicate that one variable decreases as the other variable increases
