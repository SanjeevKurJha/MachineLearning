## Bosten Housing dataset URL **[ https://archive.ics.uci.edu/ml/machine-learning-databases/housing/]**

Attribute Information:
***
    1. CRIM      per capita crime rate by town
    2. ZN        proportion of residential land zoned for lots over 25,000 sq.ft.
    3. INDUS     proportion of non-retail business acres per town
    4. CHAS      Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
    5. NOX       nitric oxides concentration (parts per 10 million)
    6. RM        average number of rooms per dwelling
    7. AGE       proportion of owner-occupied units built prior to 1940
    8. DIS       weighted distances to five Boston employment centres
    9. RAD       index of accessibility to radial highways
    10. TAX      full-value property-tax rate per $10,000
    11. PTRATIO  pupil-teacher ratio by town
    12. B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
    13. LSTAT     % lower status of the population
    14. MEDV     Median value of owner-occupied homes in $1000's***

### Model fitting 

#### Chose a class of model 
##### first we are going to fit a model with LSTAT and RM with MEDV individually for simple linear regression

#### https://scikit-learn.org/stable/modules/linear_model.html#ransac-regression
#### http://digitalfirst.bfwpub.com/stats_applet/stats_applet_5_correg.html

```python
from sklearn.linear_model import RANSACRegressor
```

#### Chose Model Hyper Parameters and instantiate modle object

```python
model=RANSACRegressor()
```

```python
model.fit(X,y)
```

```python
inlier_mask=model.inlier_mask_
outlier_mask=np.logical_not(inlier_mask)
```

#### Draw a line with number between 3 to 10 with steps of 1  and predict the value

```python
line_x=np.arange(3,10,1)
line_y_predict=model.predict(line_x.reshape(-1,1))
```

```python
plt.figure(figsize=(20,15))
plt.scatter(X[inlier_mask],y[inlier_mask],c='blue',marker='o',label="inliers")
plt.scatter(X[outlier_mask],y[outlier_mask],c='red',marker='s',label="outliers")
plt.plot(line_x,line_y_predict,color="red")
plt.xlabel("Average number of rooms per dwelling")
plt.ylabel("Median value of owner-occupied homes in $1000's")
plt.legend(loc="upper")
plt.show()
```

```python
model.estimator_.coef_
```
